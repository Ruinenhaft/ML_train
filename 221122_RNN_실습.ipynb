```python
import yfinance as yf
import pandas as pd
```

```python
history = yf.Ticker('MSFT').history(period='max')
history.head()
```

```python
# 2015-01-01 / 2020-12-31
# Date, Open, High, Low, Volume, Close
history.columns
```

```python
# Date 인덱스에서 뽑아내기
history.reset_index(inplace=True)
history
```

```python
history.drop(["Dividends", "Stock Splits"], axis=1, inplace=True)
history
```

```python
df = history[history['Date'].between('2020-01-01', '2022-10-31')].reset_index(drop=True)
df
```

```python
df.info()
```

```python
# 시간 순서대로 주가를 분석해서 -> 다음날의 종가(Close)를 예측하는 모델을 구축.
# 인풋 -> 개장가, 최고가, 최저가
# 타겟 -> 종가.
import matplotlib.pyplot as plt

df_used = df[['Open','High','Low']]
df_used['Close'] = df['Close']
hist = df_used.hist()
plt.show()
```

```python
# Dataset 학습용 데이터셋 (MS)
import numpy as np
from torch.utils.data.dataset import Dataset

class Microsoft(Dataset):
    def __init__(self):
        # 데이터 정규화
        self.data = df_used.drop('Close', axis=1).values # 값만 있는 배열
        self.data = self.data / np.max(self.data) # np.max 배열에서 가장 큰 값

        self.label = df_used['Close'].values
        self.label = self.label / np.max(self.label)
    
    # 사용가능한 배치 개수를 반환하는 __len__()
    def __len__(self):
        return len(self.data) - 30 # 사용가능한 배치 계수, 30일 기준으로 학습을 시켜줄건데, 72일부터는 30일치를 만들수없다. 
        # 30일치가 묶여서 batch 


    # 30일치 데이터를 읽어오는 함수
    def __getitem__(self, i): # 0 ~ len-1
        data = self.data[i:i+30] # 독립변수
        label = self.label[i+30] # 종속변수
        return data, label
```

```python
# RNN 모델 정의
import torch
import torch.nn as nn

class RNN(nn.Module) : 
    def __init__(self):
        super(RNN, self).__init__()

        # RNN 층 정의
        # Open, High, Low 3개 columns
        self.rnn = nn.RNN(input_size = 3,
                          hidden_size = 8,
                          num_layers = 5, # RNN층을 얼마나 쌓을지(주로 3, 5)
                          # 너무 많으면 기울기 소실, 너무 적으면 기울기 폭발
                          batch_first=True) # Batch 차원(30)

        # FC, MLP => 분류, 예측을 위한 층.
        # 주가를 예측하는 MLP층 정의, 피쳐 = 배치 * 히든사이즈
        self.fc1 = nn.Linear(in_features= 240, out_features = 64)
        self.fc2 = nn.Linear(in_features= 64, out_features = 1)

        self.relu = nn.ReLU()

    # 이전에 결과물이 소멸하는게 아니라 다음 사이클에 반영되어야 한다.
    def forward(self, x, h0):# h0 = 가장 초반에 넣었던 가중치를 의미
        x, hn = self.rnn(x,h0) # RNN층의 출력
        # x -> 층을 거쳐 나온 결과 (출력층)
        # hn -> 모든 RNN층의 은닉 상태 (누적 가중치들)

        # x.shape[0] = batch size / 두 개의 차원을 1차원으로 변환해준다.
        # MLP 층의 입력으로 사용되게 모양 변경
        x = torch.reshape(x, (x.shape[0], -1))

        # MLP 층을 사용해서 값 예측
        x = self.fc1(x)
        x = self.relu(x) # 활성화함수
        x = self.fc2(x)

        # 출력층(벡터형태), 텐서 안에 존재하기때문에
        # 평면으로 전환
        x = torch.flatten(x)

        return x
```

```python
# 학습 시작
from tqdm.notebook import tqdm

from torch.optim.adam import Adam
from torch.utils.data.dataloader import DataLoader

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = RNN().to(device)
dataset = Microsoft()
```

```python
# 데이터 로더
loader = DataLoader(dataset, batch_size=32)
```

```python
# 최적화 정의
optim = Adam(params=model.parameters(), lr = 0.0001)
```

```python
# 학습루프
for epoch in range(200):
    iterator = tqdm(loader)
    for data, label in iterator:#tqdm쓰려고 넣어준다.
        optim.zero_grad()#최초의 가중치 0
        # 초기 은닉상태
        h0 = torch.zeros(5, data.shape[0], 8).to(device)
        # 모델 예측값
        preds = model(data.type(torch.FloatTensor).to(device), h0)
        # 손실 계산
        loss = nn.MSELoss()(preds, label.type(torch.FloatTensor).to(device))# 선형모델
        # 역전파
        loss.backward() # 역전파
        optim.step() # 최적화

        iterator.set_description(f"epoch{epoch} loss:{loss.item()}")

torch.save(model.state_dict(), "./rnn.pth")
```

```python
# 시각화
import matplotlib.pyplot as plt

loader = DataLoader(dataset, batch_size =1) # 예측값을 위한 데이터 로더

preds = [] # 예측값 저장할 리스트
total_loss = 0

# 경사 안들어가게
with torch.no_grad():
    model.load_state_dict(torch.load('rnn.pth', map_location=device))

    for data, label in loader :
        h0 = torch.zeros(5, data.shape[0], 8).to(device)

        pred = model(data.type(torch.FloatTensor).to(device), h0)
        preds.append(pred.item()) # 실제값과 비교를 위해서 예측값을 저장

        # 손실 계산
        loss = nn.MSELoss()(pred, label.type(torch.FloatTensor).to(device))
        total_loss += loss/len(loader) # 손실의 평균치
```
